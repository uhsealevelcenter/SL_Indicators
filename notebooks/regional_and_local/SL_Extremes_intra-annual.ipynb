```{figure} https://uhslc.soest.hawaii.edu/jfiedler/SeaLevelIndicators/output/Hawaii_Region_Output/SeasonalExtremeVariations_1612340.png
---
height: 400px
name: seasonal_extreme_variations
---
A climatology of extreme sea levels at the Honolulu tide gauge. Monthly maxima (+), determined from hourly sea level data, are shown from 1993-2024. The solid line corresponds to projected sea level return levels at 2,10,50 and 100 years, using this 30-year chunk of data. This line uses a nonstationary GEV to take the seasonal variations of sea level into account. The 2, 10, 50 and 100-year return levels are shown to vary throughout the year (dashed line), with the lowest extremes occuring in March, and highest extremes peaking in July. No long-term trend is used in this model, and the scale and location parameters are held constant.
```


The figure below was created in the [contributions](https://jwfiedler.github.io/SL_Hawaii/notebooks/contributions.html#plot-sub-annual-contributions) notebook.

```{figure} https://uhslc.soest.hawaii.edu/jfiedler/SeaLevelIndicators/output/Hawaii_Region_Output/location_by_year_Mokuoloe.png
---
height: 400px
name: location by year
---
This plot shows the seasonal and yearly (denoted by color) changes in the location parameter (with covariate included) from 1993 through 2023 for the non-stationary GEV model at Mokuoloe. This particular calculation uses the BEST climate index at an 18-month lag. While there is a steady trend (background sea level rise), this is also affected by the covariate and nodal influences.
```
